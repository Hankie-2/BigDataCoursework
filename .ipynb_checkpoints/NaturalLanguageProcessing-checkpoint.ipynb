```python
text
```

## 

### Разделим текст на предложения

```python
from nltk.tokenize import sent_tokenize
nltk.download('punkt')
print('')

sentences = sent_tokenize(text)
sentences
```

### 

### Разделим предложения на слова

```python
from nltk.tokenize import word_tokenize

words = []

for sentence in sentences:
    t = word_tokenize(sentence)
    words.extend(t)
    
words
```

### 

### Мы должны избавиться от предлогов, союзов и знаков пунктуации, так как при дальнейшем анализе они не имеют ни какого смысла

```python
from nltk.corpus import stopwords
nltk.download('stopwords')

print('')

unique_stops = set(stopwords.words('russian'))

no_stops = []

for word in words:
    
    word = word.lower()
    
    if word not in unique_stops and word.isalpha():
        no_stops.append(word)
        
no_stops
```

### 

### Теперь нам нужно привести все слова к начальной форме 

```python
import pymorphy2
morph = pymorphy2.MorphAnalyzer()
```

```python
lemmatized = []

for word in no_stops:
    if len(word) ==1:
        continue
    p = morph.parse(word)[0]
    word = p.normal_form
    lemmatized.append(word)

print(lemmatized)
```

## 

### Рассчитываем, количество повторений слов в тексте

```python
from collections import Counter

bow_counter = Counter(lemmatized)

print(bow_counter)
```
